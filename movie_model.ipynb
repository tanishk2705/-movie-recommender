```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/movie-recommendor-ML/movie-content-recommendor ML/tmdb_5000_movies.csv')
credits=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/movie-recommendor-ML/movie-content-recommendor ML/tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies=movies.merge(credits, on='title')
```

```python
movies.shape
```

```python
movies=movies[['movie_id','title','keywords','overview','genres','cast','crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i["name"])
  return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
import ast
def convert3(obj):
 counter = 0
 L = []
 for i in ast.literal_eval(obj):
    counter += 1
    L.append(i['name'])
    if(counter==3):
      break
 return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['cast'][0]
```

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if(i['job']=='Director'):
     L.append(i["name"])
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies['crew'][0]
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  L=[]
  for i in text.split():
    L.append(ps.stem(i))

  return " ".join(L)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommendmovie(movie):
  movie_index=new_df[new_df['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key =lambda x:x[1])[1:21]  # print 9 similar movies as searched
  for i in movies_list:
   print(new_df.iloc[i[0]].title)
```

```python
recommendmovie('Avatar')
```
